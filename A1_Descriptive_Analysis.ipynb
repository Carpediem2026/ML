<p><b>Задача 2.</b> Анализ рынка акций (4 балла)</p>

<p>1. Загрузите данные в один dataframe из всех файлов в папке /data/stock. Все файлы имеют одинаковую структуру, в том числе наименование столбцов. В качестве значений индекса dataframe'а необходимо указать значения столбца "Date". Название столбцов должные соответствовать названию акций (имя файла без .csv), а их значения - значениям цены закрытия (столбец "Close" в файлах .csv)</p>

```python
# Ваш код
url1 = "https://raw.githubusercontent.com/Carpediem2026/ML/main/stock/ADBE.csv"
z2 = pd.read_csv(url1)
z2.pop('Open')
z2.pop('High')
z2.pop('Low')
z2.pop('Volume')
z2.pop('Adj Close')
z2.rename(columns = {'Close' : 'ADBE'}, inplace = True) 
z2=z2.set_index('Date')
url_all = pd.Series(["https://raw.githubusercontent.com/Carpediem2026/ML/main/stock/ADSK.csv", "https://raw.githubusercontent.com/Carpediem2026/ML/main/stock/AKAM.csv",
                     "https://raw.githubusercontent.com/Carpediem2026/ML/main/stock/BABA.csv","https://raw.githubusercontent.com/Carpediem2026/ML/main/stock/BIDU.csv",
                     "https://raw.githubusercontent.com/Carpediem2026/ML/main/stock/CA.csv","https://raw.githubusercontent.com/Carpediem2026/ML/main/stock/CHKP.csv",
                     "https://raw.githubusercontent.com/Carpediem2026/ML/main/stock/CRM.csv","https://raw.githubusercontent.com/Carpediem2026/ML/main/stock/CTSH.csv",
                     "https://raw.githubusercontent.com/Carpediem2026/ML/main/stock/EBAY.csv","https://raw.githubusercontent.com/Carpediem2026/ML/main/stock/FB.csv",
                     "https://raw.githubusercontent.com/Carpediem2026/ML/main/stock/GOOGL.csv","https://raw.githubusercontent.com/Carpediem2026/ML/main/stock/INFY.csv",
                     "https://raw.githubusercontent.com/Carpediem2026/ML/main/stock/INTU.csv","https://raw.githubusercontent.com/Carpediem2026/ML/main/stock/MSFT.csv",
                     "https://raw.githubusercontent.com/Carpediem2026/ML/main/stock/NTES.csv","https://raw.githubusercontent.com/Carpediem2026/ML/main/stock/NVDA.csv",
                     "https://raw.githubusercontent.com/Carpediem2026/ML/main/stock/PYPL.csv","https://raw.githubusercontent.com/Carpediem2026/ML/main/stock/RHT.csv",
                     "https://raw.githubusercontent.com/Carpediem2026/ML/main/stock/SAP.csv","https://raw.githubusercontent.com/Carpediem2026/ML/main/stock/TWTR.csv",
                     "https://raw.githubusercontent.com/Carpediem2026/ML/main/stock/VMW.csv","https://raw.githubusercontent.com/Carpediem2026/ML/main/stock/YHOO.csv"])
name_url = pd.Series(['ADSK','AKAM','BABA','BIDU','CA','CHKP','CRM','CTSH','EBAY','FB','GOOGL','INFY','INTU','MSFT','NTES','NVDA','PYPL','RHT','SAP','TWTR','VMW','YHOO'])
for i in range(url_all.size) :
  url2 = url_all[i]
  z2_1 = pd.read_csv(url2)
  z2_1.pop('Open')
  z2_1.pop('High')
  z2_1.pop('Low')
  z2_1.pop('Volume')
  z2_1.pop('Adj Close')
  z2_1.rename(columns = {'Close' : name_url[i]}, inplace = True) 
  z2_1=z2_1.set_index('Date')
  z2=pd.concat([z2,z2_1], axis=1, join='outer')#inner') #outer
z2
```

<p>2. Рассчитайте корреляционную матрицу для всех акций</p>

```python
# Ваш код
corr_matrix = z2.corr()
corr_matrix
```
